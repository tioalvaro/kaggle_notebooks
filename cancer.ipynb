```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Set up Complete')
```

```python
cancer_filepath = "../raw_data/cancer.csv"
```

```python
cancer_data = pd.read_csv(cancer_filepath, index_col='Id')
cancer_data.head()
```

```python
sns.histplot(data=cancer_data, x='Area (mean)', hue='Diagnosis')
```

```python
# looking at the histogram, we can see that there is a correlation between higher means areas with malign diagnosis. 
```

```python
sns.histplot(data=cancer_data, x='Texture (mean)', hue='Diagnosis')
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12,9))
plt.title("Worst Radius on Cancer Cells")
sns.kdeplot(data=cancer_data, x='Radius (worst)', hue='Diagnosis', fill=True)

#Given this data, is more likely that Radius (Worst) values higher than 25 are more likely to be a malignant tumor
```

```python
sns.jointplot(data=cancer_data, x=cancer_data['Area (mean)'], y=cancer_data['Radius (worst)'], kind="kde", hue='Diagnosis')
```
