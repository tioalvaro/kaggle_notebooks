```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
insurance_filepath = "../raw_data/insurance.csv"
insurance_data = pd.read_csv(insurance_filepath)
```

```python
insurance_data.head()
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])

# adding a linear regression to understand the trend between bmi and insurance charges
```

```python
# we can also add other factors to our analysis using color coded scatter plots

sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
plt.figure(figsize=(10,8))
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)

# lmplot gives the ability to check two tendency lines 
```

```python
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```

```python
sns.scatterplot(x=insurance_data['age'], y=insurance_data['charges'], hue=insurance_data['sex'])
```
