```python
# counting the total values of missing data on the dataframe
missing_values_count = nfl_data.isnull().sum()
```

```python
# determining the total number of cells with .shape
total_cells = np.prod(nfl_data.shape)
total_missing = missing_values_count.sum()

percent_missing = (total_missing/total_cells)
print(percent_missing)
```
